```python
# imports
import sklearn as sk 
import pandas as pd 
from sklearn.datasets import load_digits
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt 

print("the sklearn version is {}".format(sk.__version__))
print("the pandas version is {}".format(pd.__version__))

```

```python
# get the data
digits = load_digits()

print("digits data loaded of type {} and headers {}".format(type(digits), list(digits)))
```

```python
# get the featres and targets
X = digits.data
y = digits.target

print("the feature data is of shape {} and type {} and the target data is of shape {} and type {}".format(X.shape, type(X), y.shape, type(y)))
```

```python
# create an isomap with two dimensions
isomap = Isomap(n_components = 2)

# fit the model
isomap.fit(X)

# transform the features
X_projected = isomap.transform(X)

print("the transformed X has shape {} and type {}".format(X_projected.shape, type(X_projected)))
```

```python
# plot the 2 dimensions
plt.figure(isomap.__class__.__name__)
plt.scatter(X_projected[:, 0], X_projected[:, 1], c=y, edgecolors='none', alpha=0.5, cmap='jet')
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5)
plt.show()

```
