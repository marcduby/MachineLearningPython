```python
# imports
import pandas as pd
import matplotlib.pyplot as plot
```

```python
# load the concrete data example
concrete_data = pd.read_csv("Datasets/Books/PythonMlCaseStudies/concrete_data.csv")
```

```python
# see what the data is
concrete_data.info()
```

```python
# describe the data
print(concrete_data.describe())
```

```python
# rename the columns
concrete_data.columns = ['cement', 'furnace_slag', 'flay_ash', 'water', 'superplasticizer', 'c_aggregate', 'f_aggregate', 'age', 'strength']
```

```python
# see what the data is
concrete_data.info()
```

```python
# print some correlation matrices
# set the figure size
figure = plot.figure(figsize= (15, 10.5))
plot_index = 1

# create a list of all the header column elements except the last one, then loop through
for feature in list(concrete_data.columns)[:-1]:
    # 3 x 3 matrix of plots
    plot.subplot(3, 3, plot_index)
    
    # scatter plot
    plot.scatter(concrete_data[feature], concrete_data['strength'])
    
    # labels
    plot.xlabel(feature.replace('_', ' ').title(), color = 'red')
    plot.ylabel('concrete strength'.title(), color = 'cyan')
    
    # update plot index
    plot_index += 1
    
```

```python
# look at the pair wise Pearson correlation by numbers
pd.set_option('display.width', 120)
pd.set_option('precision', 3)
correlations = concrete_data.corr(method= 'pearson')


print(correlations.head(20))
```
