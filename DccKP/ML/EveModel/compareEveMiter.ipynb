```python
# read the csv
df_eve = pd.read_csv(file_eve_pparg)
```

```python
# subselect columns
df_eve_smaller = df_eve[['position', 'mt_aa', 'EVE_scores_ASM']]

# pivot
df_eve_pivot = df_eve_smaller.pivot(index='position', columns='mt_aa')

# analyze
print("got eve data \n{}".format(df_eve_pivot.tail(20)))
```
